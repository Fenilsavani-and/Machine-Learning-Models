# Gradient Descent:

```python
import numpy as np

# yhat = wx + b

x = np.random.randn(10,1)
y = 3*x + np.random.randn()

# Initialize w and b value:
w, b = 0,0
# Initialize learning rate alpha value:
alpha = 0.01

def Gradient(x, y, w, b, alpha):

    N = x.shape[0] # sample size of training
    dldw, dldb = 0,0

# dervation of loss function with w nd b: and I need to find for all sample and sum it
    dldw = sum(map(lambda xi,yi: -2*xi*(yi - (w*xi+b)), x,y))
    dldb = sum(map(lambda xi,yi:-2*(yi - (w*xi+b)), x,y))

# Make an update to the w parameter
    w = w - alpha*(1/N)*dldw
    b = b - alpha*(1/N)*dldb

    return w,b

for i in range(400):
  w,b = Gradient(x,y,w,b,alpha)
  yhat = w*x + b
  loss = np.divide(np.sum((y-yhat)**2, axis = 0), x.shape[0])
  print(f'{i} loss is {loss}, for paramters w:{w}, b:{b}')



```
